```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
df_5x5 = pd.read_csv('hitori_heuristics_metrics_5x5.csv')
df_5x5.head()
```

```python
average_time = df_5x5['Time'].mean()
max_time = df_5x5['Time'].max()
min_time = df_5x5['Time'].min()

print(f"Average Time: {average_time} (seconds)")
print(f"Maximum Time: {max_time} (seconds)")
print(f"Minimum Time: {min_time} (seconds)")
average_function_calls = df_5x5['FunctionCalls'].mean()
max_function_calls = df_5x5['FunctionCalls'].max()
min_function_calls = df_5x5['FunctionCalls'].min()
average_function_calls_log = math.ceil(math.log2(average_function_calls)*100)/100
max_function_calls_log = math.ceil(math.log2(max_function_calls)*100)/100
min_function_calls_log = math.ceil(math.log2(min_function_calls)*100)/100

print(f"Average Function Calls: {average_function_calls} ~ 2^{average_function_calls_log}")
print(f"Maximum Function Calls: {max_function_calls} ~ 2^{max_function_calls_log}")
print(f"Minimum Function Calls: {min_function_calls} ~ 2^{min_function_calls_log}")
```

```python
df_6x6 = pd.read_csv('hitori_heuristics_metrics_6x6.csv')
df_6x6.head()
```

```python
average_time = df_6x6['Time'].mean()
max_time = df_6x6['Time'].max()
min_time = df_6x6['Time'].min()

print(f"Average Time: {average_time} (seconds)")
print(f"Maximum Time: {max_time} (seconds)")
print(f"Minimum Time: {min_time} (seconds)")
average_function_calls = df_6x6['FunctionCalls'].mean()
max_function_calls = df_6x6['FunctionCalls'].max()
min_function_calls = df_6x6['FunctionCalls'].min()
average_function_calls_log = math.ceil(math.log2(average_function_calls)*100)/100
max_function_calls_log = math.ceil(math.log2(max_function_calls)*100)/100
min_function_calls_log = math.ceil(math.log2(min_function_calls)*100)/100

print(f"Average Function Calls: {average_function_calls} ~ 2^{average_function_calls_log}")
print(f"Maximum Function Calls: {max_function_calls} ~ 2^{max_function_calls_log}")
print(f"Minimum Function Calls: {min_function_calls} ~ 2^{min_function_calls_log}")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df_5x5['FunctionCalls'], df_5x5['Time'])
plt.title('Function Calls vs Time')
plt.xlabel('Function Calls')
plt.ylabel('Time')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df_6x6['FunctionCalls'], df_6x6['Time'])
plt.title('Function Calls vs Time')
plt.xlabel('Function Calls')
plt.ylabel('Time')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Difficulty', y='FunctionCalls', data=df_5x5)
plt.title('Boxplot of Function Calls for each Difficulty Level')
plt.xlabel('Difficulty')
plt.ylabel('Function Calls')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Difficulty', y='FunctionCalls', data=df_6x6)
plt.title('Boxplot of Function Calls for each Difficulty Level')
plt.xlabel('Difficulty')
plt.ylabel('Function Calls')
plt.show()
```

```python
metrics = ['Time', 'FunctionCalls']
df_list = [df_5x5, df_6x6]
labels = ['5x5', '6x6']

for metric in metrics:
    means = [df[metric].mean() for df in df_list]
    maxs = [df[metric].max() for df in df_list]
    mins = [df[metric].min() for df in df_list]

    x = np.arange(len(labels))  # the label locations
    width = 0.2  # the width of the bars

    fig, ax = plt.subplots()
    rects1 = ax.bar(x - width, means, width, label='Mean')
    rects2 = ax.bar(x, maxs, width, label='Max')
    rects3 = ax.bar(x + width, mins, width, label='Min')

    ax.set_ylabel(metric)
    ax.set_title(f'Comparison of {metric} between 5x5 and 6x6')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend()

    fig.tight_layout()

    plt.show()
```
